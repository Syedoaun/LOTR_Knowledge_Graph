```python
import pandas as pd
import string

# Assuming df1 is your DataFrame

# Define a function to remove punctuation, including parentheses, from a string
def remove_punctuation_and_parentheses(text):
    translation_table = str.maketrans("", "", string.punctuation + "(" + ")" + "[\"']" )
    return text.translate(translation_table)

# Apply the function to all string columns in the DataFrame
df1 = df1.applymap(lambda x: remove_punctuation_and_parentheses(str(x)) if isinstance(x, str) else x)

```

#### Assigning Gender

```python
# set the gender
with onto10:
    for i in df1['name']:
        char = onto10.Character(i)
        if (df1[df1['name'] == i]['gender'].iloc[0] == "Female"):
            char.hasgender.append("Female")
        elif (df1[df1['name'] == i]['gender'].iloc[0] == "Male"):
            char.hasgender.append("Male")
```

```python
# has name

with onto10:
    for i in df1['name']:
        onto10.Name(i)
    for j in df1['realm'].unique():
        onto10.Name(j)
    for k in df1['race1'].unique():
        onto10.Name = k

```

#### Character hasrace Race

```python
# belongs to race
with onto10:
    for i in df1['name']:
        char = onto10.Character(i)
        race = df1[df1['name'] == i]['race1'].iloc[0]
        race_class = None
        try:
            race_class = onto.search_one(iri='*'+ race)
        except Exception:
            pass
        if race_class:
            char.hasrace.append(race_class)

```

#### Character dateofbirth

```python
with onto10:
    for i in df1['name']:
        char = onto10.Character(i)
        dob = df1[df1['name'] == i]['birth'].iloc[0]
        print(f"Processing {i}, Date of Birth: {dob}")
        char.dateofbirth = dob

```

#### Character dateofdeath

```python
with onto10:
    for i in df1['name']:
        char = onto10.Character(i)
        dod = df1[df1['name'] == i]['death'].iloc[0]

        print(f"Processing {i}, Date of death: {dod}")
        char.dateofdeath = dod
```

#### Character spouceof Character

```python
with onto10:
    for i in df1['name']:
        char = onto10.Character(i)
        spouse = df1[df1['name'] == i]['spouse'].iloc[0]

        print(f"Processing {i}, spouse: {spouse}")

        # Check if dod is not None and is a valid individual in the ontology
        #print(onto[dod],"llllllllllllll")
        if spouse is not None and onto10[spouse] is not None:
            char.spouceof.append(onto10[spouse])
        else:
            # Handle the case where dod is None or not found in the ontology
            print('p')
            pass
```
